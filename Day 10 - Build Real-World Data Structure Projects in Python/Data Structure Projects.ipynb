## **INVENTORY MANAGEMENT SYSTEM**
---
- The Inventory Management System provides a basic framework for managing products, categories, and stock levels using Python's data structures effectively.
- The code is using three data structures for storing product we are using dictionaries, for dtoring the ordera list is used and for storing the categories a set is used as no duplicates are allowed in it.
- All the related functions are implemented that include the addition, updation and display of the products in the inventory.

```python
# Define data structures
products = {}      # Dictionary 
product_order = []     # List
categories = set()      # Set
```

```python
# Functions
def add_product(product_id, name, category, stock):
    if product_id not in products:
        products[product_id] = {'name': name, 'category': category, 'stock': stock}
        product_order.append(product_id)
        categories.add(category)
        print(f"The product {name} with ID: {product_id} has been added to category {category} with stock {stock}.")
    else:
        print(f"Product ID {product_id} already exists.")

def update_stock(product_id, quantity):
    if product_id in products:
        products[product_id]['stock'] += quantity
        print(f"Updated stock for product {products[product_id]['name']} (ID: {product_id}) to {products[product_id]['stock']}.")
    else:
        print(f"Product ID {product_id} not found.")

def display_products():
    print("\nAll Products:")
    for pid in product_order:
        product = products[pid]
        print(f"ID: {pid}, Name: {product['name']}, Category: {product['category']}, Stock: {product['stock']}")
    print()

def display_by_category(category):
    print(f"\nAll products in Category '{category}':")
    for pid in product_order:
        product = products[pid]
        if product['category'] == category:
            print(f"ID: {pid}, Name: {product['name']}, Stock: {product['stock']}")
    print()
```

```python
# Testing
add_product('01', 'Table', 'Furniture', 50)
add_product('02', 'Chair', 'Furniture', 30)
add_product('03', 'Pencil', 'Stationary', 20)
print("\n")
update_stock('01', 40)
update_stock('02', 5)

display_products()

display_by_category('Furniture')
```

## **PHONE BOOK APPLICATION**
---
- The phone book application is a simple program that will be saving the names of person and their phone numbers.
- The program is implemented with te help of Dictionaries as it provides key-value pairs that can be helpful in saving the phone numbers against the names of person.

Below is the code for the complete application and it has different functions lik adding contact, deleting contact, updating an existing contacts.

```python
# Phone book dictionary to store contacts
phone_book = {}

def add_contact(name, phone_number):
    if name in phone_book:
        print(f"{name} is already in the phone book.")
    else:
        phone_book[name] = phone_number
        print(f"Added {name} with phone number {phone_number} to the phone book.")

def delete_contact(name):
    if name in phone_book:
        del phone_book[name]
        print(f"Deleted {name} from the phone book.")
    else:
        print(f"{name} is not found in the phone book.")

def update_contact(name, new_phone_number):
    if name in phone_book:
        phone_book[name] = new_phone_number
        print(f"Updated phone number for {name} to {new_phone_number}.")
    else:
        print(f"{name} is not found in the phone book.")

def search_contact(name):
    if name in phone_book:
        print(f"Phone number for {name}: {phone_book[name]}")
    else:
        print(f"{name} is not found in the phone book.")

def display_all_contacts():
    print("\nPhone Book:")
    for name, phone_number in phone_book.items():
        print(f"{name}: {phone_number}")
    print()

# Example usage:
add_contact("Tehreem", "03332456365")
add_contact("Ali", "0336567677")
add_contact("Sara", "0336789454")

display_all_contacts()

update_contact("Ali", "0333456466")
delete_contact("Sara")

display_all_contacts()

search_contact("Ali")
search_contact("Sara")
```

**Now that we have learned the OOP basics in python so a more sophisticated approach would be to create a class of Phone book that has the attribute of the dictionary and methods of adding , deleting and updating contacts.**

So a revised version of te above code is given below.

```python
class PhoneBook:
    def __init__(self):
        self.contacts = {}

    def add_contact(self, name, phone_number):
        if name in self.contacts:
            print(f"{name} is already in the phone book.")
        else:
            self.contacts[name] = phone_number
            print(f"Added {name} with phone number {phone_number} to the phone book.")

    def delete_contact(self, name):
        if name in self.contacts:
            del self.contacts[name]
            print(f"Deleted {name} from the phone book.")
        else:
            print(f"{name} is not found in the phone book.")

    def update_contact(self, name, new_phone_number):
        if name in self.contacts:
            self.contacts[name] = new_phone_number
            print(f"Updated phone number for {name} to {new_phone_number}.")
        else:
            print(f"{name} is not found in the phone book.")

    def search_contact(self, name):
        if name in self.contacts:
            print(f"Phone number for {name}: {self.contacts[name]}")
        else:
            print(f"{name} is not found in the phone book.")

    def display_all_contacts(self):
        print("Phone Book:")
        for name, phone_number in self.contacts.items():
            print(f"{name}: {phone_number}")
        print()
```

```python
# Example usage with user input:
phone_book = PhoneBook()

while True:
    print("\nPhone Book Menu:")
    print("1. Add Contact")
    print("2. Delete Contact")
    print("3. Update Contact")
    print("4. Search Contact")
    print("5. Display All Contacts")
    print("6. Exit")

    choice = input("Enter your choice (1-6): ")

    if choice == '1':
        name = input("Enter name: ")
        phone_number = input("Enter phone number: ")
        phone_book.add_contact(name, phone_number)
    elif choice == '2':
        name = input("Enter name to delete: ")
        phone_book.delete_contact(name)
    elif choice == '3':
        name = input("Enter name to update: ")
        new_phone_number = input("Enter new phone number: ")
        phone_book.update_contact(name, new_phone_number)
    elif choice == '4':
        name = input("Enter name to search: ")
        phone_book.search_contact(name)
    elif choice == '5':
        phone_book.display_all_contacts()
    elif choice == '6':
        print("Exiting Phone Book. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 6.")
```

## **RESTAURANT MENU SYSTEM**
---
The Restaurant menu system is a platform where customers can view the menu, place orders, and calculate the total bill. 
- Contains key-value pairs where keys are menu items (like 'burger', 'pizza') and values are their respective prices.
- Order is stored in dictionaries where each dictionary represents an item in the order with keys 'item' and 'quantity'.
- Then we have several functions for displaying menu items, add item to order , and calculating bill.
- The code also coontains a complete menu that allows user to select the option he/she wants to avail.

```python
# Define the restaurant menu using a dictionary with prices in rupees
menu = {
    'burger': 440,
    'pizza': 660,
    'salad': 475,
    'fries': 183,
    'drink': 146,
    'pasta': 805,
    'sandwich': 586,
    'soup': 366,
    'steak': 1174,
    'chicken': 917,
    'taco': 293,
    'sushi': 733,
    'burrito': 623,
    'pancake': 403,
    'ice cream': 256
}
```

```python
# List to store customer orders
orders = []
```

```python
# Function to display the restaurant menu
def display_menu():
    print("\n*****MENU ITEMS*****")
    for item, price in menu.items():
        print(f"{item.capitalize()}: Rs. {price:.2f}")
    print()
```

```python
# Function to add items to customer's order
def add_to_order(item, quantity):
    if item in menu:
        orders.append({'item': item, 'quantity': quantity})
        print(f"Added {quantity} {item}(s) to your order.")
    else:
        print(f"Sorry! {item.capitalize()} is not available in the menu.")
```

```python
# Function to calculate the total bill
def calculate_bill():
    total_bill = 0.0
    for order in orders:
        item = order['item']
        quantity = order['quantity']
        if item in menu:
            total_bill += menu[item] * quantity
    return total_bill
```

```python
# Main program
def main():
    print("\n\n\t*****Welcome to the Restaurant Menu System!*****")
    
    while True:
        print("\n\nWhat would you like to do?")
        print("1. Display Menu")
        print("2. Add Item to Order")
        print("3. View Order")
        print("4. Calculate Total Bill")
        print("5. Exit")
        choice = input("Enter your choice (1-5): ")
        
        if choice == '1':
            display_menu()
        elif choice == '2':
            item = input("Enter the item you want to add to your order: ").lower()
            quantity = input_quantity(f"How many {item}(s) would you like to order? ")
            add_to_order(item, quantity)
        elif choice == '3':
            if orders:
                print("Your Order:")
                for order in orders:
                    print(f"{order['item'].capitalize()}: {order['quantity']}")
            else:
                print("Your order is empty.")
        elif choice == '4':
            total_bill = calculate_bill()
            if total_bill > 0:
                print(f"Your total bill is: Rs. {total_bill:.2f}")
            else:
                print("Your order is empty. Total bill is Rs. 0.00")
        elif choice == '5':
            print("Thank you for visiting. Have a great day!")
            break
        else:
            print("Invalid choice. Please enter a number from 1 to 5.")

def input_quantity(prompt):
    while True:
        try:
            quantity = int(input(prompt))
            if quantity <= 0:
                print("Please enter a positive number.")
                continue
            return quantity
        except ValueError:
            print("Invalid input. Please enter a valid number.")

if __name__ == "__main__":
    main()
```
