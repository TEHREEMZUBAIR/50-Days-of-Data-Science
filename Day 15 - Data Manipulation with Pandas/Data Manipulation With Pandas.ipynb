## **INDEXING**

Indexing in Pandas is a fundamental skill that allows you to efficiently access and manipulate data. This guide will cover various indexing techniques for both Series and DataFrame objects.

---
### **SERIES INDEXING**
We can access elements using labels or positions:

```python
# From a list
data = [10, 20, 30, 40, 50]
index = ['a', 'b', 'c', 'd', 'e']
series = pd.Series(data, index=index)
```

```python
# accessing a single element
value = series['a']
value
```

```python
# accessing multiple elememts
value = series[['a', 'd', 'e']]
value
```

```python
# slicing by label
slice = series['b' : 'd']
slice
```

```python
# slicing by position
slice = series [1 : 4]
slice
```

```python
# Conditional slicing
filtered = series[series > 25]
filtered
```

```python
# dataframe creation
df = pd.read_csv('/kaggle/input/countries-population/country_db.csv')
```

```python
# printing first five rows
df.head(10)
```

```python
# setting the df column as index
df.set_index('country', inplace=True)
df.head()
```

```python
# displaying population column
population_column = df['population']
population_column
```

```python
# selecting multiple columns
subset_df = df[['population', 'continent']]
subset_df.head()
```

```python
# Select Subset of Rows Using .loc
albania_data = df.loc['Albania']
albania_data
```

```python
# Select Subset of Rows and Columns Using .iloc
subset = df.iloc[:3, :2]
subset
```

```python
# Filter Rows Based on a Condition
high_population_countries = df[df['population'] > 40000000]
print("\nCountries with population greater than 40 million:")
high_population_countries
```

```python
# Group by a Specific Column and Calculate Mean
mean_population_by_continent = df.groupby('continent')['population'].mean()
mean_population_by_continent.head()
```

```python
# Group by Multiple Columns and Calculate Sum
sum_population_by_continent_year = df.groupby(['continent', 'year'])['population'].sum()
print("\nSum of population by continent and year:")
sum_population_by_continent_year
```

```python
#  Apply Multiple Aggregation Functions Using agg
agg_functions = {
    'population': ['mean', 'sum']
}
agg_result = df.groupby('continent').agg(agg_functions)
agg_result
```

```python
# Calculate Size of Each Group
group_sizes = df.groupby('continent').size()
group_sizes
```

```python
# Select Rows Based on Multiple Conditions
filtered_df = df[(df['population'] > 30000000) & (df['number_of_languages'] < 5)]
filtered_df
```

```python
# Use the query Method to Filter Rows
filtered_query = df.query('population > 100000000')
print("\nFiltered DataFrame using query method (population > 100 million):")
filtered_query
```

```python
# Use isin to Filter Rows Based on a List of Values
filtered_isin = df[df['continent'].isin(['Europe', 'Asia'])]
print("\nFiltered DataFrame using isin method (Europe or Asia):")
filtered_isin
```

```python
# Select Specific Columns and Rename Them
selected_columns = df[['year', 'population']].rename(columns={'population': 'total_population'})
print("\nSelected columns with renamed 'population' to 'total_population':")
selected_columns
```
