## **1. LIST**

## **2. TUPLE**
- Tuple are a collection of python objects much like a list but tuples are immutable in nature i.e the elements in the tuple cannot be added or removed once created.
- Just like a List, a Tuple can also contain elements of various types.

```python
# creating a tuple
book = ("Python Programming", "John Doe", 2023)
print("Original tuple:", book)
```

```python
# accessing elements
title = book[0]
author = book[1]
year = book[2]
print("Title:", title)
print("Author:", author)
print("Year:", year)

# using negative index
print("Negative indexing")
title = book[-3]
author = book[-2]
year = book[-1]
print("Title:", title)
print("Author:", author)
print("Year:", year)
```

```python
# Tuple unpacking
title, author, year = book
print("Unpacked values:", title, author, year)
```

```python
# Length of tuple
length_of_tuple = len(book)
print("Length of book tuple:", length_of_tuple)
```

```python
# Checking membership
is_python_book = "Python Programming" in book
print("Is 'Python Programming' in book?", is_python_book)
```

```python
# Finding index of an element
index_of_year = book.index(2023)
print("Index of 2023 in book:", index_of_year)
```

```python
# checking immutable nature
# Trying to modify a tuple (will raise an error)
try:
    book[0] = "Advanced Python Programming"
except TypeError as e:
    print("Cannot modify tuple:", e)
```

```python
# tuple concatenation
additional_info = ("Beginner's Guide", "Education Company")
extended_book_info = book + additional_info
print("Concatenated tuple:", extended_book_info)
```

```python
# repeatig tuple
repeated_book_info = book * 2
print("Repeated tuple:", repeated_book_info)
```

## **3. SETS**
- Python Set is an unordered collection of data.
- It is mutable and does not allow any duplicate element.
- Sets are basically used to include membership testing and eliminating duplicate entries.
- The data structure used in this is Hashing, a popular technique to perform insertion, deletion, and traversal in O(1) on average. 

```python
# set creation

# method 1
set1 ={1, 2, 3, 4, 5}
set2 = set([3, 4, 5, 6, 7])

# all entries should be unique
```

```python
# add element to a set
set1 ={1, 2, 3, 4, 5}
set1.add(6)  # Adds element 6 to set1
print(set1)

# shows that sets are mutable
```

```python
# removing an element
set1 ={1, 2, 3, 4, 5, 6}
set1.remove(6)  # Adds element 6 to set1
print(set1)
```

```python
# checking for duplicate insertion
my_set = {1, 2, 3, 4, 5}
print("Before adding duplicates: ", my_set)
# Attempting to add a duplicate element
my_set.add(3)  # This will not add 3 again since sets do not allow duplicates
print("After attempting to add 3 again:", my_set)
```

```python
# union of two sets
set1 = {1, 2, 3, 4, 5}
set2 = {"apple", "mango", "banana", "orange"}

union_set = set1 | set2
print(union_set)

# using union method
print("Using union method..")
union_set_method = set1.union(set2)
print(union_set_method)
```

```python
# intersection of two sets
set1 = {1, 2, 3, 4, 5, "mango"}
set2 = {1, 2, "apple", "mango", "banana", "orange"}
intersection_set = set1 & set2
print(intersection_set)

# using intersection method
print("Using Intersection method..")
intersection_set_method = set1.intersection(set2)
print(intersection_set_method)
```

```python
# difference of two sets
set1 = {1, 2, 3, 4, 5, "mango"}
set2 = {1, 2, "apple", "mango", "banana", "orange"}
difference_set = set1 - set2
print(difference_set)

# using difference method
print("Using Difference method..")
difference_set_method = set1.difference(set2)
print(difference_set_method)
```

```python
# some other operations
set1 = {1, 2, "mango"}
set2 = {1, 2, "apple", "mango", "banana", "orange"}

is_subset = set1 <= set2
print("Is set1 a subset of set2?", is_subset)
is_superset = set1 >= set2
print("Is set1 a superset of set2?", is_superset)
is_proper_subset = set1 < set2
print("Is set1 a proper subset of set2?", is_proper_subset)
is_proper_superset = set1 > set2
print("Is set1 a proper superset of set2?", is_proper_superset)
```

```python
# set comprehension
set3 = {x for x in range(10) if x % 2 == 0}
print("Set created using comprehension:", set3)
```

## **4. DICTIONARY**
- Python dictionary is like hash tables in any other language with the time complexity of O(1).
- It is an unordered collection of data values, used to store data values like a map.
- Dictionary holds the key:value pair. Key-value is provided in the dictionary to make it more optimized. 
- Indexing of Python Dictionary is done with the help of keys. 

```python
# Creating dictionaries
person = {
    'name': 'Alice',
    'age': 30,
    'city': 'New York'
}

# Another way using dict() constructor
car = dict(
    brand='Toyota',
    model='Camry',
    year=2022
)
```

```python
# accessing values in a dictionary
print(person)
print(person['name'])  # Output: 'Alice'
```

```python
# adding a new element
person = {
    'name': 'Alice',
    'age': 30,
    'city': 'New York'
}

person['email'] = 'alice@example.com'  # Adding a new key-value pair

# updating existing element
person['age'] = 31  # Updating the value for the 'age' key

print(person)
```

```python
car = dict(
    brand='Toyota',
    model='Camry',
    year=2022
)

# deleting elements
del car['year']  # Deleting the 'year' key
removed_value = car.pop('brand')  # Removing 'city' key and retrieving its value

print(car)
```

```python
# DICTIONARY METHODS
person = {
    'name': 'Alice',
    'age': 30,
    'city': 'New York'
}

print(person.keys())
print(person.values())
print(person.items())
```

```python
# get method
# Safely retrieve a value by key, with a default if key not found.

person = {
    'name': 'Alice',
    'city': 'New York'
}
age = person.get('age', 'N/A')  # If 'age' key exists, return its value; otherwise, return 'N/A'
print(age)
```

```python
# update method
# Merge another dictionary into the current dictionary.
person = {
    'name': 'Alice',
    'city': 'New York',
    'age' : 30
}

additional_info = {
    'phone': '123-456-7890',
    'email': 'alice@gmail.com'
}

person.update(additional_info)
print(person)
```

```python
# iterate through dictionary
person = {
    'name': 'Alice',
    'city': 'New York',
    'age' : 30
}

print("ITERATING THROUGH KEYS")
for key in person:
    print(key, ':', person[key])

print("ITERATING THROUGH KEY-VALUE PAIR")
for key, value in person.items():
    print(key, ':', value)
```

```python
# dictionary comprehension
numbers = {x: x**2 for x in range(1, 9)}
print(numbers)
```

## **FROZEN SETS**
- A frozenset in Python is an immutable version of a set.
- Once created, its elements cannot be changed or modified. 
- Frozensets are useful when you need a set-like data structure that is hashable and can be used as a key in dictionaries or as an element in another set.

**USE CASES**

Frozen sets are particularly useful in scenarios where:
   
   - You need a collection of unique elements that should not be changed accidentally.
   - You want to use a set as a key in a dictionary or as an element in another set.

```python
# creating frozen set
# Creating frozen sets
fs1 = frozenset([1, 2, 3, 4, 5])
fs2 = frozenset(['apple', 'banana', 'cherry'])
```

```python
# Operations on frozen sets
fs1 = frozenset([1, 2, 3, 4, 5])
fs2 = frozenset([4, 5, 6, 7, 8])

union_fs = fs1 | fs2
intersection_fs = fs1 & fs2
difference_fs = fs1 - fs2

print("Union of fs1 and fs2:", union_fs)
print("Intersection of fs1 and fs2:", intersection_fs)
print("Difference of fs1 and fs2:", difference_fs)
```

```python
# Methods and properties
fs = frozenset([1, 2, 3, 4, 5])

# Methods
print("Length of frozen set fs:", len(fs))  # Length of the frozen set
print("Minimum element in fs:", min(fs))    # Minimum element in the frozen set
print("Maximum element in fs:", max(fs))    # Maximum element in the frozen set
```

## **BYTE ARRAY**
- Bytearrays in Python are mutable sequences of bytes (integers ranging from 0 to 255).
- They provide a way to manipulate binary data efficiently and are especially useful when you need to modify bytes in place.

**USE CASES**

Bytearrays are particularly useful in scenarios where:

  - You need to modify binary data efficiently.
  - You want to work with mutable sequences of bytes, such as when reading from or writing to binary files.
  - You need to perform operations like appending, inserting, or deleting elements in the byte sequence.

```python
# Creating bytearrays
ba1 = bytearray(b'hello')  # Using bytes literal
ba2 = bytearray([65, 66, 67, 68, 69])  # Using list of integers (ASCII values)
```

```python
# Basic operations
# Creating a bytearray
ba = bytearray(b'hello')

# Modifying bytearray
ba[0] = 106  # ASCII value for 'j'

# Concatenating bytearray
ba.extend(b' world')

print("Modified bytearray:", ba)
```

```python
# Methods and properties
ba = bytearray(b'hello')

# Methods
ba.append(33)  # Append ASCII value for '!'
ba.insert(5, 32)  # Insert ASCII value for space at index 5
ba.pop()  # Remove and return the last item

print("Updated bytearray:", ba)
print("Length of bytearray:", len(ba))
```

## **STACKS**
Stacks are a fundamental data structure in computer science that follows the Last In, First Out (LIFO) principle.

- We can implement stacks using lists.
- To add an item to the top of the stack, use append(). 
- To retrieve an item from the top of the stack, use pop() without an explicit index. For example:

```python
# Initialize a stack with elements 3, 4, and 5
stack = [3, 4, 5]

# Add element 6 to the top of the stack
stack.append(6)

# Add element 7 to the top of the stack
stack.append(7)

# Print the current stack
print(stack)  # Output: [3, 4, 5, 6, 7]

# Remove and return the top element of the stack (LIFO)
print(stack.pop())  # Output: 7

# Print the stack after popping
print(stack)  # Output: [3, 4, 5, 6]

# Remove and return the top element of the stack (LIFO)
print(stack.pop())  # Output: 6

# Remove and return the top element of the stack (LIFO)
print(stack.pop())  # Output: 5

# Print the stack after popping
print(stack)  # Output: [3, 4]
```

## **QUEUES**
- It is also possible to use a list as a queue, where the first element added is the first element retrieved (“first-in, first-out”).
- While appends and pops from the end of list are fast, doing inserts or pops from the beginning of a list is slow (because all of the other elements have to be shifted by one).
- To implement a queue, use collections.deque which was designed to have fast appends and pops from both ends. For example:

```python
from collections import deque

# Initialize a deque named queue with initial elements
queue = deque(["Eric", "John", "Michael"])

# Print the initial queue
print("Initial queue:", queue)  # Output: Initial queue: deque(['Eric', 'John', 'Michael'])

# Terry arrives and joins the queue
queue.append("Terry")
print("Queue after Terry arrives:", queue)  # Output: Queue after Terry arrives: deque(['Eric', 'John', 'Michael', 'Terry'])

# Graham arrives and joins the queue
queue.append("Graham")
print("Queue after Graham arrives:", queue)  # Output: Queue after Graham arrives: deque(['Eric', 'John', 'Michael', 'Terry', 'Graham'])

# The first to arrive (Eric) now leaves
first_to_leave = queue.popleft()
print("First person to leave:", first_to_leave)  # Output: First person to leave: Eric

# The second to arrive (John) now leaves
second_to_leave = queue.popleft()
print("Second person to leave:", second_to_leave)  # Output: Second person to leave: John

# Print the remaining queue in order of arrival
print("Remaining queue:", queue)  # Output: Remaining queue: deque(['Michael', 'Terry', 'Graham'])
```
