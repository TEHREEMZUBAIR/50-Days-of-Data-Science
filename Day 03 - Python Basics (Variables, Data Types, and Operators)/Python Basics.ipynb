# **CALCULATE MONTHLY ELECTRICITY BILL BASED ON THE NUMBER OF UNITS CONSUMED**

This code is calculating the electricity bill based on the number of units consumed by the customer. 
* First it takes the number of units consumed by the customer as input.
* After getting number of units consumed the bill is calculated.
* If units consumed are greater than 100 than the rate is 3.5, else the rate is 5.00
* Also a fixed charge of 50 is added to the billing amount.

```python
# define electricity rates
RATE_UPTO_100_UNITS = 3.50
RATE_ABOVE_100_UNITS = 5.00
FIXED_CHARGE = 50.00

units_consumed = float(input("ENTER THE NUMBER OF UNITS CONSUMED THIS MONTH: "))

if units_consumed <= 100:
    bill_amount = (units_consumed * RATE_UPTO_100_UNITS) + FIXED_CHARGE
else:
    bill_amount = (units_consumed * RATE_ABOVE_100_UNITS) + FIXED_CHARGE

print("YOUR ELECTRICITY BILL FOR THIS MONTH IS Rs.", bill_amount)
```

# **CALCULATE THE GPA OF SUBJECT BASED ON STUDENT ATTENDANCE, PRE-TERMINAL EXAM AND TERMINAL EXAM MARKS**

### Bullet Point Description of the Code

- **Input Attendance**: The program prompts the user to enter the student's attendance percentage.
- **Check Attendance**: If the attendance is 80% or higher:
  - **Input Marks**: The program asks for the pre-terminal and terminal marks (out of 50 each).
  - **Calculate Total Marks**: The sum of pre-terminal and terminal marks is computed.
  - **Determine GPA and Grade**: Based on the total marks, the GPA and corresponding grade are assigned using a series of conditional statements:
    - `>= 80`: GPA 4.0, Grade A+
    - `>= 75`: GPA 3.67, Grade A
    - `>= 70`: GPA 3.33, Grade B+
    - `>= 65`: GPA 3.00, Grade B
    - `>= 60`: GPA 2.67, Grade B-
    - `>= 55`: GPA 2.33, Grade C+
    - `>= 50`: GPA 2.0, Grade D
    - `< 50`: GPA 1.0, Grade FAIL
  - **Output GPA and Grade**: The GPA and grade are printed.
- **Short Attendance**: If the attendance is less than 80%, a message stating the student is debarred from the subject due to short attendance is printed.

```python
try:
    attendance = int(input("ENTER THE ATTENDANCE PERCENTAGE OF STUDENT (0-100): "))
    if attendance < 0 or attendance > 100:
        raise ValueError("Attendance should be between 0 and 100.")
    
    if attendance >= 80:
        try:
            pre_term_marks = int(input("ENTER THE PRE-TERMINAL MARKS OUT OF 50: "))
            if pre_term_marks < 0 or pre_term_marks > 50:
                raise ValueError("Pre-terminal marks should be between 0 and 50.")
            
            final_marks = int(input("ENTER THE TERMINAL MARKS OF STUDENT OUT OF 50: "))
            if final_marks < 0 or final_marks > 50:
                raise ValueError("Terminal marks should be between 0 and 50.")
            
            total_marks = pre_term_marks + final_marks
            
            if total_marks >= 80:
                gpa = 4.0
                grade = 'A+'
            elif total_marks >= 75:
                gpa = 3.67
                grade = 'A'
            elif total_marks >= 70:
                gpa = 3.33
                grade = 'B+'
            elif total_marks >= 65:
                gpa = 3.00
                grade = 'B'
            elif total_marks >= 60:
                gpa = 2.67
                grade = 'B-'
            elif total_marks >= 55:
                gpa = 2.33
                grade = 'C+'
            elif total_marks >= 50:
                gpa = 2.0
                grade = 'D'
            else:
                gpa = 1.0
                grade = 'FAIL'
            
            print(f"YOUR GPA FOR THIS COURSE IS: {gpa}/4 AND GRADE IS: {grade}")
        
        except ValueError as e:
            print(f"Invalid input: {e}")
    else:
        print("THE STUDENT HAS BEEN DEBARRED FROM THE SUBJECT DUE TO SHORT ATTENDANCE")

except ValueError as e:
    print(f"Invalid input: {e}")
```
