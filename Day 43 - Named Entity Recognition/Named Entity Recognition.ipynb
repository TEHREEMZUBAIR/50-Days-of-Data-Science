# **Loading and Tokenizing the Text**
Next we need to define our sample text on which we want to the named entity recognition

```python
sample_text = 'On July 20, 1969, Neil Armstrong and Buzz Aldrin became the first humans to set foot on the Moon. This historic Apollo 11 mission, conducted by NASA, was watched by millions around the world. The lunar module, named Eagle,landed in the Sea of Tranquility. During their mission, Armstrong and Aldrin conducted experiments and collected samples. The event was a major milestone in the Space Race between the United States and the Soviet Union, a key component of Cold War history. The success of the mission was celebrated in Washington, D.C., and had a profound impact on global space exploration.'
```

Next we need to tokenize our text. For this we have already studied various tokenization techinques.

Here we will  be using word tokenization, as we want to acheive NER on the wod level of the text.

```python
tokens = nltk.word_tokenize(sample_text)
print(tokens)
```

# **Performing Part-of-Speech (POS) Tagging**
After tokenizing the text, we will perform part-of-speech tagging on the tokens to understand the role of each word in the sentence.

For this we will be using NLTK pos_tag function. Here is the code:

```python
tags = nltk.pos_tag(tokens)

print(tags)
```

Here is a brief description of what each pos tag means.
- **JJ:** Adjective (describes a noun)
- **NN, NNS:** Noun, singular/plural (person, place, thing)
- **NNP:** Proper noun, singular (name of a person, organization)
- **VBD, VBG, VB:** Different verb tenses (past, gerund, base)
- **DT:** Determiner (the, a)
- **CD:** Cardinal number (5.1, billion)
- **IN:** Preposition (on, in, for)
- **PRP$:** Possessive pronoun (its)

Now once we have done POS tagging it is now easier for us to perform NER on the data, as now we know the structure of our text and the meaning of our text. Now we will perform NER on the tags.

# **Performing Named Entity Recognition (NER)**'
Now we need to perforn the NER on the tags that we produced before.

NLTK prvides and extensive function named `ne_chunk()` that performs the NER task.
- The main task of this function is to identifiy the named entities in the tags and then classify them into categories such as: persons, organizations, locations etc.

```python
entities = nltk.ne_chunk(tags)
print(entities)
```

- **European:** Tagged as a GPE (Geopolitical Entity), representing a geographical region.
- **Google:** Mistakenly tagged as a PERSON, but should have been tagged as an ORGANIZATION.
- **Menlo Park:** Correctly tagged as a GPE (if present in the text), but not part of this particular output.

# **Extracting and Displaying Named Entities**
We will extract the recognized named entities from the chunked tree structure and display them.


Here's what the code does, without showing the lines of code: 

- Construction of empty list to store named entities and their types. 
- An empty list is created to store the named entities and their types.
- Iteration: For each element in the tree generated by the NER process, the code checks to see whether it is a node rather than a tip. 
- In short, if the element is a subtree, this provides evidence that it represents a named entity. 
- Extraction: For each subtree to be specific, It collects every word that belongs to the named entity and turns these into a string representing the whole named entity. 
- Then it gets the label of the subtree, which denotes the kind of named entity this represents (e.g., PERSON, ORGANIZATION, GPE). 
- Storage: With this information, the extracted named entity and its type are put in the list for later use. Output: Finally, the list of named entities along with their types is printed.

```python
# Extract named entities
named_entities = []
for subtree in entities:
    if isinstance(subtree, nltk.Tree):
        entity = " ".join([word for word, tag in subtree.leaves()])
        entity_type = subtree.label()
        named_entities.append((entity, entity_type))
# Print extracte entities
print(named_entities)
```

So that was a complete demonstration of how we can do Named Entity Recognition using NLTK. But the problem with NLTK is that it does not provide any visualization tools for a better understanding.

<h1 style="color: black; font-size: 50px; font-weight: bold;">SENTIMENT ANALYSIS OF TWEETS</h1>

Sentiment Analysis is a content rewriter, known in addition as opinion mining, it's a natural language processing (NLP) technology used to determine the sentiment behind a series of words and then analyze the emotional orientation. It can assist people to understand the feelings revealed by text data, which can be divided into many different emotions such as positive or negative with nuance in between.

## **Key Concepts in Sentiment Analysis**

#### **1. Text Data:** 
The main input to sentiment analysis is text data, which comes from many places including social media posts, reviews, news articles and other text-based content.

#### **2. Sentiment Classification:** 
The aim here is to put the text into predefined categories of mood. Some common categories are:

- **Positive:** Text in favor of a favourable view.
- **Negative:** Text opposed to an unfavourable POV.
- **Neutral:** Text which doesn't express strong feelings one way or the other, and so could be seen as both positive and negative in equal measure.

#### **3. Subjectivity and Polarity:**
- **Subjectivity:** How subjective or opinionated the text is. A higher subjectivity indicates that the text is more opinion-based. 
- **Polarity:** The degree of sentiment in the text, ranging from negative to positive. It quantifies sentiment on a scale, usually from -1 (most negative) to +1 (most positive). 

#### **4. Techniques and Tools:**
- **TextBlob:** A library that offers a simple API for doing common natural language processing (NLP) tasks, including sentiment analysis. 
- **NLTK (Natural Language Toolkit):** A powerful library for working with human language data, among other tasks it includes sentiment analysis using machine learning algorithms. 
- **SpaCy:** An advanced NLP library capable of various NLP tasks, including named entity recognition and sentiment analysis.

Here is complete guide to how we can perform sentiment analysis.

# **Cleaning the Text Data**
Next we need to clean our data. For these we need to do following steps:

**1. @mentions:** Remove any mentioned Twitter handle in the tweet. These kind of twitter handles for example @Ali are removed bacuause they donot contribute in the detemining the sentimnent if the tweet.


```python
df['text'] = df['text'].apply(lambda text: re.sub(r'@[A-Za-z0-9]+', '', text))
```

**2. Hashtags:** The # symbol used in hashtags (e.g., #GOPDebate) is removed. Hashtags are often not meaningful for sentiment analysis as they are not part of the core message of the tweet.


```python
df['text'] = df['text'].apply(lambda text: re.sub(r'#', '', text))
```

**3. RT:** The RT indicator signifies that the tweet is a retweet. Removing it helps in focusing on the content of the original tweet rather than its retweet status.

```python
df['text'] = df['text'].apply(lambda text: re.sub(r'RT[\s]+', '', text))
```

**4. Hyperlinks:** 
Hyperlinks are removed as they do not provide useful information for sentiment analysis and can introduce noise into the dataset.

```python
df['text'] = df['text'].apply(lambda text: re.sub(r'https?:\/\/\s+', '', text))
```

**5. Colons:**
Colons are often used in various contexts that may not be relevant to sentiment analysis. Removing them helps in normalizing the text.

```python
df['text'] = df['text'].apply(lambda text: re.sub(r':+', '', text))
```

**6. Dashes:**
Dashes are also removed to avoid potential noise. They may be used for various purposes in tweets, but they don't contribute to sentiment analysis.

```python
df['text'] = df['text'].apply(lambda text: re.sub(r'--+', '', text))
```


**7. 'http' substring:**
The substring 'http' is removed to clean up any remaining fragments of URLs that may have been missed by previous steps.

```python
df['text'] = df['text'].apply(lambda text: re.sub(r'http', '', text))
```

```python
df.head()
```

# **Extracting Subjectivity and Polarity**
Now we need to compute the subjectivity and polarity of our tweets.

### **Polarity** 
- Polarity measures how positive or negative a piece of text is. - It ranges from -1 (very negative) to +1 (very positive). A polarity score of 0 indicates neutral sentiment.
- Polarity helps in understanding the sentiment of the text, which is useful for tasks like sentiment analysis, customer feedback evaluation, and social media monitoring.

### **Subjectivity** 
- Subjectivity measures how subjective or objective a piece of text is. 
- It ranges from 0 (very objective) to 1 (very subjective). A subjectivity score close to 1 means the text is more of an opinion or personal experience, while a score close to 0 means it is more factual or objective.
- Subjectivity helps in distinguishing between factual content and personal opinions, which can be useful for content analysis, summarizing opinions, and understanding the nature of the text.

We can use TextBlob to compute polarity and subjectivity, here is how we can do that:

```python
df['subjectivity'] = df['text'].apply(lambda text: TextBlob(text).sentiment.subjectivity)
```

```python
df.head()
```

```python
df['polarity'] = df['text'].apply(lambda text: TextBlob(text).sentiment.polarity)
df.head()
```

# **Preparing Test Data**
This section further cleans the test text data by removing hashtags, mentions, URLs, and stopwords. It prepares the text for feature extraction.

In this section we will remove the stop words from the test data 

```python

text = []
stopwords_set = set(stopwords.words("english"))

for index, row in train.iterrows():
    words_filtered = [e.lower() for e in row.text.split() if len(e) >= 3]
    words_without_stopwords = [word for word in words_filtered if not word in stopwords_set]
    text.append((words_without_stopwords, row.sentiment))
```

# **Feature Extraction**
We define functions to extract features from the text, creating a feature set based on the most common words in the dataset.

```python
# Extracting word features
def get_words_in_text(text):
    all_words = []
    for (words, sentiment) in text:
        all_words.extend(words)
    return all_words
```

```python
def get_word_features(wordlist):
    wordlist = nltk.FreqDist(wordlist)
    features = wordlist.keys()
    return features
w_features = get_word_features(get_words_in_text(text))
```

```python
def extract_features(document):
    document_words = set(document)
    features = {}
    for word in w_features:
        features['contains(%s)' % word] = (word in document_words)
    return features
```

# **Training the Classifier**
We will be using Naive Bayes Classifier using the extracted features from the tarining set.

```python
training_set = nltk.classify.apply_features(extract_features,text)
classifier = nltk.NaiveBayesClassifier.train(training_set)
```

# **Evaluating the Classifier**
We evaluate the classifier's performance on the test data, calculating the number of correctly classified positive and negative tweets.

```python
neg_cnt = 0
pos_cnt = 0
for obj in test_neg: 
    res =  classifier.classify(extract_features(obj.split()))
    if(res == 'Negative'):
        neg_cnt = neg_cnt + 1
for obj in test_pos:
    res =  classifier.classify(extract_features(obj.split()))
    if(res == 'Positive'):
        pos_cnt = pos_cnt + 1
        
print('[Negative]: %s/%s '  % (len(test_neg),neg_cnt))       
print('[Positive]: %s/%s '  % (len(test_pos),pos_cnt))
```

There seems to be a good performance in predicting Negative sentiment, as a high number of predictions (1594/1649) for Negative were correct.
However, for Positive sentiment, the model seems to have more difficulty, as only 138/456 predictions were correctly classified as Positive, indicating a higher number of misclassifications for the Positive class.

- The model might be biased towards predicting Negative sentiment more often, which could be due to class imbalance (more Negative examples in the dataset) or the model being less sensitive to Positive instances.
- Further adjustments such as balancing the dataset, tuning hyperparameters, or trying a different model could help improve Positive sentiment classification.
